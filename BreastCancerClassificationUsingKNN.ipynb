```python
import numpy as np
import math
from collections import Counter
from sklearn import datasets
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
```

```python
def euclideanDist(a,b):
  return np.sqrt(np.sum((a-b)**2))

class KNN:
  def __init__(self, k):
    self.k = k

  def fit(self,X,y):
    self.X_train = X
    self.y_train = y

  def predict(self,X):
    predictions = []
    for x in X:
      predictions.append(self.predictHelper(x))
    return predictions

  def predictHelper(self,x):
    distances = []
    for x_t in self.X_train:
      distances.append(euclideanDist(x, x_t))

    nearest_indices = np.argsort(distances)[:self.k]
    k_nearest_labels = []
    for i in nearest_indices:
      k_nearest_labels.append(self.y_train[i])


    most_common = Counter(k_nearest_labels).most_common(5)

    return most_common[0][0]
```

```python
cmap = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
```

```python
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target # 0 is malignant tumor, 1, is benign tumor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, edgecolor='k', s=20)
plt.show()
```

```python
clf = KNN(k = int(math.sqrt(len(X_train))))
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
print (predictions)

accuracy = np.sum(predictions == y_test) / len(y_test)
print(accuracy)
```
